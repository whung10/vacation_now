```python
# take San Francisco (SFO) for example and read all the price history; define a function to read flight data

filenames = ['~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/ATL_SFO.json',
             '~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/BOS_SFO.json',
             '~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/BWI_SFO.json',
             '~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/CLT_SFO.json',
             '~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/DCA_SFO.json']
#flight_dataset = {'flight_labels': ['monthly_flight_avg', 'monthly_flight_min', 'month'], 'flight_data': []}

def flight(filenames):
    """Get monthly average flight ticket price of each city and minimum price"""
    for f in filenames:
        print(f[-12:-5])
        
        df = pd.read_json(f)
        df_avg = df.groupby(['year']).mean()        
        print(df_avg)
        
        df_min = df_avg[df_avg['price'] == df_avg['price'].min()]
        print(df_min)
```

```python
flight(filenames)
```

```python
# load weather API for historical weather data; take San Francisco (aiport code: SFO) for example

url = 'https://api.meteostat.net/v1/history/monthly?station=72494&start=2018-01&end=2018-12&key=Ku9EsoIm'

r = requests.get(url)
json_data = r.json()
df = pd.DataFrame.from_dict(json_data['data'])
df.set_index('month')
print(df)
```

```python
# visualization 

df['raindays'].plot(kind='bar', color='b')
plt.xlabel('Month')
plt.ylabel('Raindays (days)')
plt.title('Raindays per month 2018')
plt.show()

df['temperature_mean'].plot(kind='line', marker='o')
plt.xlabel('Month')
plt.ylabel('Average Temperature (C)')
plt.title('Average Temperature per month 2018')
plt.show()

df_diff = df['temperature_mean_max'] - df['temperature_mean_min']
df_diff.plot(kind='line', marker='o', color='r')
plt.xlabel('Month')
plt.ylabel('Average Temperature Change (C)')
plt.title('Average Temperature Change per month 2018')
plt.show()
```

```python
# Select the best month to travel

def find_best_month(city):
    """return the best month to travel to this city"""
    
    url = 'https://api.meteostat.net/v1/history/monthly?station='+city+'&start=2018-01&end=2018-12&key=Ku9EsoIm'
    r = requests.get(url)
    json_data = r.json()
    df = pd.DataFrame.from_dict(json_data['data'])
    df.set_index('month')
    
    temp_mean = df['temperature_mean']
    for t in temp_mean:
        if t <= 25 and t >= 16:
            print(df[df['temperature_mean'] == t])
    
    rdays = df['raindays']
    for r in rdays:
        if r == r.min:
            print(df[df['raindays'] == r])
            
find_best_month('72494')
```
